## Section1 機械学習とは

### 1.1 AI? ディープラーニング?

**機械学習**は、人工知能研究の1分野、**ディープラーニング**は機械学習の1手法のことを指します。


<img src="figures/ai-ml-dl.png" width="60%">

#### 人工知能 （AI: Artificial Intelligence）

人間と同等（もしくはそれ以上）の知能を機械に再現させる取り組み・概念のことです。
1950年頃に第一次ブームが起こったらしい。

*keyword:* `ルールベース`


#### 機械学習 （ML: Machine Learning）

機械が大量のデータから規則性やルールを見つけ出す手法です。
入力するデータは人間が特徴量を抽出する必要があります。

*keyword:* `ランダムフォレスト`、`K-means`、`サポートベクターマシン（SVM）`

#### ニューラルネットワーク （NN: Neural Network）

脳神経細胞（ニューロン）のしくみを参考にした機械学習の一種です。
基本的に教師あり学習ですが、ディープラーニングの他分野応用にともなって利用範囲が拡大しています。[議論有]

#### ディープラーニング （DL: Deep Learning）

ニューラルネットワークを多層化することで発展した技術で、入力されたデータから特徴量を自ら抽出することが可能です。

*keyword:* `畳み込みニューラルネットワーク（CNN）`、`リカレントニューラルネットワーク（RNN）`、 `敵対的生成ネットワーク（GAN）`、`深層強化学習`

### 1.2 特徴量

対象の特徴が数値化されたもののことを **特徴量 (Feature value)** と言います。

機械学習では複数の特徴量から学習を行います。この特徴量の数を **次元 (Dimension)** と言います。

### 1.3 機械学習の分類

機械学習ではデータに潜む構造・パターンを捉えるという大目標の下、データの扱い方によって 3 つに分類されます。

#### 教師あり学習 (Supervised learning)

事前にラベル付けされたデータをもとに、入出力関係を捉える手法です。株価予測や画像認識によく使われます。

されに、出力が離散的なら **分類 (Classification)** 、連続的なら **回帰 (Regression)** と区別します。

|                            | 分類                       | 回帰                 |
| -------------------------- | -------------------------- | -------------------- |
| 出力                       | 離散的（カテゴリ）           | 連続的（数値）       |
| 利用シーン                 | 画像認識<br>迷惑メール判定 | 株価予測<br>異常検知 |
| 例）<br>気温,湿度,風速から | 天気（雨,晴れ,曇り）を予測 | 降水量を予測         |
| 代表的なアルゴリズム | 決定木<br>ロジスティクス回帰<br>ニューラルネットワーク | 線形回帰<br>Ridge回帰<br>Lasso回帰 |


#### 教師なし学習 (Unsupervised learning)

ラベル付けされていない（出力すべきものが決まっていない）データからデータ自身の構造や分布を捉えます。画像生成や異常検知に使われています。

具体的な例として **クラスター分析 (Clustering)** 、 **主成分分析 (PCA)** などがあります。


#### 強化学習 (Reinforcement learning)

状態・行動・報酬からなるデータの対応関係を捉えることで、報酬を最大化するような行動を特定できます。囲碁やテレビゲームの対戦型AIによく使われます。

### 1.4 機械学習の流れ

機械学習モデルを実際に運用するまでの流れを大枠で捉えると、

1. **データの収集・前処理**
2. **モデルの検討・構築**
3. **学習**
4. **性能評価**
5. **運用**

というようになります。

### 1.5 学習とは

教師あり学習における学習とは、**用意した入出力データを用いて、モデルが同じデータを再現するように** 訓練することです。

訓練 = 「同じ入力に対する出力データとモデルの出力の"ズレ度合い"を下げる」

ということです。
